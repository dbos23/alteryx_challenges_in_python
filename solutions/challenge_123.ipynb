```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
# set date and human population. they'll increment each month

date = datetime(2018, 7, 2)
human_pop = 326766748

# create df

data = [{
    'MalePop': 1,
    'FemalePop': 1,
    'Age': 0,
}]

for i in range(1, 24):
    data.append({
        'MalePop': 0,
        'FemalePop': 0,
        'Age': i
        })

df = pd.DataFrame(data)

df
```

```python
# define function to count all the rabbits in the df

def count_rabbits(df):

    agg_df = df.agg({
        'MalePop': 'sum',
        'FemalePop': 'sum'
    }).to_frame().T

    return int(agg_df.loc[0, 'MalePop'] + agg_df.loc[0, 'FemalePop'])
```

```python
while count_rabbits(df) < 2 * human_pop:
    # calculate the number of babies born
    adult_females = int(df[df['Age'] >= 4]['FemalePop'].sum())
    babies = pd.DataFrame([{
        'MalePop': adult_females,
        'FemalePop': adult_females * 3,
        'Age': 0
    }])
    
    # increment date, human pop, and age of rabbits, dropping dead rabbits
    date += relativedelta(months=1)
    human_pop = round(human_pop * 1.00058)
    df['Age'] += 1
    df = df[df['Age'] < 24]

    # add in babies
    df = pd.concat([babies, df], ignore_index=True)
```

```python
print(f'Rabbit Population: {count_rabbits(df)}')
print(f'Human Population: {human_pop}')
print(f'Date: {date}')
```
