```python
import pandas as pd
```

```python
# create dataframe for starting data

data = []

for i in range(1, 1001):
    dict = {}
    dict['locker'] = i
    dict['position'] = 0
    data.append(dict)

df = pd.DataFrame(data)
df.head()
```

```python
for student in range(1, 1001):
    for i in range(0, 1000):
        if round(df.loc[i, 'locker'] / student) == (df.loc[i, 'locker'] / student):
            df.loc[i, 'position'] = df.loc[i, 'position'] + 1
```

```python
for i in range(1000):
    if round(df.loc[i, 'position'] / 2) == (df.loc[i, 'position'] / 2):
        df.loc[i, 'pos'] = 'Closed'
    else:
        df.loc[i, 'pos'] = 'Open'
```

```python
df = df[df['pos'] == 'Open']
df.head()
```
