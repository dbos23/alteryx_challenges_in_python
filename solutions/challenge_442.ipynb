```python
import pandas as pd
```

```python
df = pd.read_csv('../data/challenge_442.csv')

df.head()
```

```python
# Output 1: calculate the yards gained by pass and run each year and the pct of total yards that came from runs

df = df.query('play_type in ("run", "pass")')
df['year'] = df['game_date'].apply(lambda x: x[:4])

df_1 = df.groupby(['year', 'play_type'])['yards_gained'].sum().reset_index()
df_1 = pd.pivot(df_1, index='year', columns='play_type', values='yards_gained').reset_index()

df_1['pct run'] = round((df_1['run'] / (df_1['pass'] + df_1['run'])) * 100, 2)

df_1.head()
```

```python
# Output 2: identify the 3 teams that have relied most heavily on the run game

df_2 = df[df['play_type'] == 'run'].groupby('possession_team')['play_type'].count().reset_index(name='count_run_plays')

df_2.sort_values('count_run_plays', ascending=False).reset_index()[:3]
```
