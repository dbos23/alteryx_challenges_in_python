```python
import pandas as pd
```

```python
df = pd.read_csv('../data/challenge_422.csv')

df.head()
```

```python
df.info()
```

```python
# filter to relevant data

df['dt'] = pd.to_datetime(df['dt'], format='%Y-%m-%d')
df = df[df['dt'] >= '1950-01-01']

df.head()
```

```python
# Output 1: Cities with avg temperatures >= 25

df_agg = df.groupby(['City', 'Country'])['AverageTemperature'].mean().reset_index()
df_agg = df_agg[df_agg['AverageTemperature'] >= 25]

df_agg.head()
```

```python
# Output 2: Counting the country with the most cities in the above output

df_agg['Count'] = df_agg.groupby('Country')['City'].transform('count')
idx = df_agg['Count'].idxmax()

df_agg.loc[idx].to_frame().T[['Country', 'Count']]
```

```python
# Output 3: Years with the highest and lowest avg temp

df['Year'] = df['dt'].dt.year
df = df.groupby('Year')['AverageTemperature'].mean().reset_index()

idx_1 = df['AverageTemperature'].idxmin()
idx_2 = df['AverageTemperature'].idxmax()

print('Lowest:')
print(df.loc[idx_1].to_frame().T)
print('\nHighest:')
print(df.loc[idx_2].to_frame().T)
```
