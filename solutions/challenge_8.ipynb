```python
import pandas as pd
```

```python
#create dataframes from CSVs
data_df = pd.read_csv('../data/challenge_8_point_of_sale.csv')
lookup_df = pd.read_csv('../data/challenge_8_lookup_table.csv')
```

```python
#convert string to date, filter to relevant dates
data_df['Date'] = pd.to_datetime(data_df['Date'])
data_df = data_df[data_df['Date'] >= pd.to_datetime('2013-07-01')]
```

```python
#join data with lookup table
df = pd.merge(data_df, lookup_df, how='inner', on='Desc')
df.head()
```

```python
#concatenate all items per order
df = df.groupby('TicketID')['Type'].agg('|'.join).reset_index()
df.head()
```

```python
#create a filtered df with only meal deal eligible orders
meal_deal_df = df[((df['Type'].str.contains('Pizza')) | (df['Type'].str.contains('Burger'))) & (df['Type'].str.contains('Side')) & (df['Type'].str.contains('Drink'))]
```

```python
#count total orders and those eligible for the meal offer
total_count = len(df)
meal_deal_count = len(meal_deal_df)
```

```python
#calculate pct of orders eligible for meal deal
meal_deal_count / total_count
```
