```python
import pandas as pd
```

```python
df = pd.read_csv('../data/challenge_390.csv')
df.head()
```

```python
#filter to USA
df = df[df['Team'] == 'U.S.A.']
```

```python
#replace non-numeric characters and convert to a numeric column
df['runs_parsed'] = pd.to_numeric(df['Runs'].str.replace('[^0-9]', '', regex=True))
```

```python
#count the number of runs per player
df = df.groupby('Player_Name')['runs_parsed'].agg('sum').reset_index()
```

```python
#take the top 5 players by runs scored
df.sort_values('runs_parsed', ascending=False)[:5]
```
