```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../data/challenge_2.csv')

df
```

```python
# parse out fields and convert string to date

df['poem'] = df['Field_1'].apply(lambda x: re.search('"(.+?)",(.+?),.(.+?).$', x).group(1))
df['poem_id'] = df['Field_1'].apply(lambda x: re.search('"(.+?)",(.+?),.(.+?).$', x).group(2))
df['date'] = df['Field_1'].apply(lambda x: re.search('"(.+?)",(.+?),.(.+?).$', x).group(3))

df['date'] = pd.to_datetime(df['date'], format='%d-%b-%y')

df
```
