```python
import pandas as pd
from datetime import datetime
```

```python
input_df = pd.read_csv('../data/challenge_126.csv')
input_df['datetime_finished'] = pd.to_datetime(input_df['datetime_finished'])

input_df.head()
```

```python
# create df of all dates in time range

dates = pd.date_range(start='2017-10-01', end=datetime.now())
dates_df = pd.DataFrame(dates, columns=['date'])

dates_df.head()
```

```python
# create df of all countries in the data

countries = input_df['country_name'].unique()
countries_df = pd.DataFrame(countries, columns=['country'])

countries_df.head()
```

```python
# cross join dates and countries dfs

scaffolded_df = pd.merge(dates_df, countries_df, how='cross')

scaffolded_df.head()
```

```python
# join main data with scaffolded dates

df = pd.merge(scaffolded_df, input_df, how='left', left_on=['date', 'country'], right_on=['datetime_finished', 'country_name'])[['date', 'country', 'passed_exams']].sort_values(['country', 'date']).reset_index(drop=True)
df['passed_exams'] = df['passed_exams'].fillna(0)


df.head()
```

```python
# calculate running total

df['running total'] = 0

for i in range(len(df)):
    if i == 0:
        df.loc[i, 'running total'] = df.loc[i, 'passed_exams']
    elif df.loc[i, 'country'] == df.loc[i-1, 'country']:
        df.loc[i, 'running total'] = df.loc[i-1, 'running total'] + df.loc[i, 'passed_exams']

df.head()
```
