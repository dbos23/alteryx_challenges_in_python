```python
import pandas as pd
```

```python
df = pd.read_csv('../data/challenge_435.csv')
df.head()
```

```python
#assign point values per medal
df.loc[df['Medal'] == 'Gold', 'points'] = 3
df.loc[df['Medal'] == 'Silver', 'points'] = 2
df.loc[df['Medal'] == 'Bronze', 'points'] = 1
```

```python
#output 1

#aggregate
df_output_1 = df.groupby(['Team', 'Sport'])['points'].sum().reset_index()

#identify top sports per team
idx = df_output_1.groupby('Team')['points'].idxmax()
df_output_1.loc[idx]
```

```python
#create concatenated event and medal field
df_output_2 = df
df_output_2['results'] = df['Event'] + ' (' + df['Medal'] + ')'
```

```python
#aggregate df, taking total points and a concatenated field with the events and medal results
df_output_2 = df_output_2.groupby(['Team', 'Name']).agg(
    points=('points', 'sum'),
    results=('results', lambda x: ', '.join(x))
).reset_index()
```

```python
#output 2

#identify top athlete per team
idx = df_output_2.groupby('Team')['points'].idxmax()
df_output_2.loc[idx]
```
