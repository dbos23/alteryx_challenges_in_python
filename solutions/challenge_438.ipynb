```python
import pandas as pd
```

```python
df = pd.read_csv('../data/challenge_438.csv')

df.head()
```

```python
# unpivot to put the dates into a column. convert it to a date and sort on it

df = pd.melt(df, id_vars=['Subgroup Name', 'Order', 'Currency'])
df['variable'] = pd.to_datetime(df['variable'])

df = df.sort_values(['Order', 'variable'])

df.head()
```

```python
# pivot so each date has its own column

df = df.pivot(columns=['variable'], values=['value'], index=['Subgroup Name', 'Order', 'Currency']).reset_index()

df
```
