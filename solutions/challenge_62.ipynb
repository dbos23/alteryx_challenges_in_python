```python
import pandas as pd
```

```python
df = pd.read_csv('../data/challenge_62.csv')

df.head()
```

```python
# convert string to datetime, sort df

df['EventFireDate'] = pd.to_datetime(df['EventFireDate'], format='%m/%d/%Y %H:%M:%S')
df = df.sort_values(['DeviceID', 'IDFA', 'EventFireDate']).reset_index(drop=True)

df.head()
```

```python
# identify duplicate clicks

df.loc[0, 'DuplicateClick'] = False

for i in range(1, len(df)):
    if (df.loc[i, 'DeviceID'] == df.loc[i-1, 'DeviceID']) and (df.loc[i, 'IDFA'] == df.loc[i-1, 'IDFA']):
        df.loc[i, 'DuplicateClick'] = (df.loc[i, 'EventFireDate'] - df.loc[i-1, 'EventFireDate']).total_seconds() <= 30
    else:
        df.loc[i, 'DuplicateClick'] = False

df
```
